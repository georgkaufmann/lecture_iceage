In this notebook, we solve a **coupled initial-value problem** numerically, using the `integrate` package from `scipy`,
and play with `python` to plot the decay time series.

We first initialize the `python` libraries.

```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
from ipywidgets import interactive
import ipywidgets as widgets
```

----
## Define a problem

As problem, we use a **mass** $m$ [kg] thrown in the vertical direction $z$. On the mass, two forces act:

- **gravity:** $F_g=mg$, with $g$ [m/s$^2$] gravitational acceleration, andf $F_g$ [Pa], gravity.
- **friction:** $F_r=k|v|v$, with $v$ [m/s] velocity, $k$ [kg/m] friction parameter, and $F_r$ [Pa] frictional force.

<img src="images/wurf1D.jpg" style="width:6cm">

These two forces are balanced, thus we can formulate an **equation of motion**:
$$
m \frac{dv}{dt} = -m g - k|v|v
$$
with $t$ time.

The velocity $v$ is defined as gradient of the vertical distance $z$:
$$
\frac{dz}{dt} = v
$$
We have **two coupled differentual equations of first order**.

**Note:** We simplified the problem to one dimension!

----
## Analytical solution

If we neglect friction, $k=0$, the two differential equations simplify to:
$$
\begin{array}{rcl}
\frac{dv}{dt} &=& -g \\
\frac{dz}{dt} &=& v
\end{array}
$$
We can solve these two differential equations analytically:
$$
\begin{array}{rcl}
v(t) &=& -g t + c_1 \\
z(t) &=& -\frac{1}{2} g t^2 + c_1  + c_2
\end{array}
$$
The two **integration constants** need to be spezified by **initial conditions**, e.g.
$v(t=0)=v_0$ and $z(t=0)=0$, thus:
$$
\begin{array}{rcl}
c_1 &=& v_0 \\
c_2 &=& 0
\end{array}
$$

As initial conditions, we use
$$
\begin{array}{rcl}
v(t=0) &=& v_0 \\
z(t=0) &=& 0
\end{array}
$$
with $g=9.81$ m/s$^2$ and $v_0=20$ m/s.

```python
g  = 9.81
v0 = 20.

t  = np.linspace(0,4,101)

v = -g*t + v0
z = -0.5*g*t**2 + v0*t
```

```python
plt.xlabel('Time [s]')
plt.plot(t,v,lw=2,ls='solid',color='red',label='v [m/s], k='+str(k[0]))
plt.plot(t,z,lw=2,ls='solid',color='blue',label='z [m],k='+str(k[0]))
plt.grid()
plt.legend()
```

----
## Numerical solution

Considering friction ($k \ne 0$), we need to solve the two differential equations numerically.
$$
\begin{array}{rcl}
\frac{dv}{dt} &=& -g - \frac{k}{m} |v| v \\
\frac{dz}{dt} &=& v
\end{array}
$$

```python
def rhs(t,y,k,m,g):
    v = y[0]
    z = y[1]
    vabs = np.abs(v)
    dvdt = -g -k/m*vabs*v
    dzdt = v
    return [dvdt,dzdt]
```

```python
m = 0.5
g = 9.81
k = np.array([0.,0.005])
yinitial = [20.,0.]
t = np.linspace(0,4,101)
```

```python
sol0 = scipy.integrate.solve_ivp(rhs,[t.min(),t.max()],yinitial,t_eval=t,args=(k[0],m,g,))
sol1 = scipy.integrate.solve_ivp(rhs,[t.min(),t.max()],yinitial,t_eval=t,args=(k[1],m,g,))

print(sol1)
```

```python
plt.xlabel('Time [s]')
plt.plot(sol0.t,sol0.y[0],lw=2,ls='solid',color='red',label='v [m/s], k='+str(k[0]))
plt.plot(sol0.t,sol0.y[1],lw=2,ls='solid',color='blue',label='z [m] ,k='+str(k[0]))
plt.plot(sol1.t,sol1.y[0],lw=2,ls='dashed',color='red',label='v [m/s], k='+str(k[1]))
plt.plot(sol1.t,sol1.y[1],lw=2,ls='dashed',color='blue',label='z [m] ,k='+str(k[1]))
plt.grid()
plt.legend()
```
