In this notebook, we solve the differential equations for **single decay** and the **decay series** with numerical methods.

```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
from ipywidgets import interactive
import ipywidgets as widgets
```

----
## Decay series

For the **different isotopes** $P_i$, we define the temporal evolution with 
**coupled initial-value differential equations of first order**:
$$
\begin{array}{rcl}
\frac{dP_1}{dt} &=& - \lambda_1 P_1 \\
\frac{dP_2}{dt} &=& + \lambda_1 P_1 - \lambda_2 P_2 \\
\vdots && \\
\frac{dP_n}{dt} &=& + \lambda_{n-1} P_{n-1}
\end{array}
$$
with $\lambda_i$ [1/s] the **decay rates**.
We use the **half-lifes** $T^i_{\frac{1}{2}}=\frac{\ln 2}{\lambda_i}$ [s] as control parameter.

```python
def rhs2(t,y,thalf1,thalf2):
    lambda1 = np.log(2)/thalf1
    lambda2 = np.log(2)/thalf2
    P1 = y[0]
    P2 = y[1]
    P3 = y[2]
    dP1dt = -lambda1*P1
    dP2dt = -lambda2*P2 +lambda1*P1
    dP3dt = +lambda2*P2
    return [dP1dt,dP2dt,dP3dt]
```

```python
# define half-live
thalf = [1.,7.]
# define time series as array
t = np.linspace(0,50,101)
# define initial conditions
yinitial = [100,0,0]
```

```python
sol2 = scipy.integrate.solve_ivp(rhs2,[t.min(),t.max()],yinitial,t_eval=t,args=(thalf[0],thalf[1],))
```

```python
plt.figure(figsize=(10.0, 6.0))
plt.title('Decay series')
plt.xlim([0,50])
plt.xlabel('Time [h]')
plt.ylim([-5,105])
plt.ylabel('Amount [%]')
plt.plot(sol2.t,sol2.y[0],linewidth=3,color='red',label='P$_1$(t)')
plt.plot(sol2.t,sol2.y[1],linewidth=3,color='green',label='P$_2$(t)')
plt.plot(sol2.t,sol2.y[2],linewidth=3,color='blue',label='P$_3$(t)')
plt.legend(loc='center right')
plt.grid()
```
