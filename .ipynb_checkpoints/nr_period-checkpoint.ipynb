```python
def nr_period(x,y,n):
    # subroutine period from Numerical Recipes
    import numpy as np
    ofac  = 4.
    hifac = 2.
    pn    = int(0.5*ofac*hifac*n)
    print ('n: ',n,' pn: ',pn)
    px = np.zeros(pn); py = np.zeros(pn)
    wpr = np.zeros(n); wpi = np.zeros(n)
    wr = np.zeros(n); wi = np.zeros(n)
    ave = np.mean(y)
    var = np.var(y)
    xmin = x.min()
    xmax = x.max()
    xdif = xmax - xmin
    xave = 0.5*(xmax+xmin)
    pymax = 0.
    pnow  = 1. / (xdif*ofac)
    for i in range(n):
        arg    = 2.*np.pi*((x[i]-xave)*pnow)
        wpr[i] = -2. * np.sin(0.5*arg)**2
        wpi[i] = np.sin(arg)
        wr[i]  = np.cos(arg)
        wi[i]  = wpi[i]
    for i in range(pn):
        px[i] = pnow
        sumsh = 0.
        sumc  = 0.
        for j in range(n):
            c = wr[j]
            s = wi[j]
            sumsh = sumsh + s*c
            sumc  = sumc + (c-s)*(c+s)
        wtau  = 0.5*np.arctan2(2.*sumsh,sumc)
        swtau = np.sin(wtau)
        cwtau = np.cos(wtau)
        sums  = 0.
        sumc  = 0.
        sumsy = 0.
        sumcy = 0.
        for j in range(n):
            s  = wi[j]
            c  = wr[j]
            ss = s*cwtau - c*swtau
            cc = c*cwtau + s*swtau
            sums = sums + ss**2
            sumc = sumc + cc**2
            yy   = y[j] - ave
            sumsy = sumsy + yy*ss
            sumcy = sumcy + yy*cc
            wtemp = wr[j]
            wr[j] = (wr[j]*wpr[j] - wi[j]*wpi[j]) + wr[j]
            wi[j] = (wi[j]*wpr[j] + wtemp*wpi[j]) + wi[j]
        py[i] = 0.5*(sumcy**2/sumc + sumsy**2/sums) / var
        if (py[i] >= pymax):
            pymax = py[i]
            jmax  = i
        pnow = pnow + 1./(ofac*xdif)
    expy = np.exp(-pymax)
    effm = 2.*pn/ofac
    prob = effm*expy
    if (prob > 0.01):
        prob = 1. - (1.-expy)**effm
    return px,py,pn
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
import ipywidgets as widgets

# define frequencies
f1 = 100000.; a1 = 1.0 # eccentricity
f2 =  41000.; a2 = 0.5 # obliquity
f3 =  23000.; a3 = 1.0 # precession
f4 =  19000.; a4 = 1.0 # precession
# timing
tmin  = -800000
tmax  = 0
tstep = 1000
rate  = 1 / tstep
nmax  = int((tmax-tmin) / tstep)+1
#-----------------------------------------------------------------------
# create time series
#-----------------------------------------------------------------------
time = np.zeros(nmax);sin0 = np.zeros(nmax)
sin1 = np.zeros(nmax);sin2 = np.zeros(nmax)
sin3 = np.zeros(nmax);sin4 = np.zeros(nmax)
i=0
for t in np.arange(tmin,tmax+tstep,tstep):
    time[i] = t
    sin1[i] = a1*np.sin(2.*np.pi*t/f1)
    sin2[i] = a2*np.sin(2.*np.pi*t/f2)
    sin3[i] = a3*np.sin(2.*np.pi*t/f3)
    sin4[i] = a4*np.sin(2.*np.pi*t/f4)
    i = i + 1
    
def plot_spectra(f1,f2,f3,f4):
    fig=plt.figure(figsize=(10.0, 6.0))
    plot1=plt.subplot2grid((2,2), (0,0), colspan=2,rowspan=1)
    plot2=plt.subplot2grid((2,2), (1,0), colspan=2,rowspan=1)
    plot1.set_xlabel('Time B.P. [y]')
    plot1.set_ylabel('Amplitude')
    sin = sin0
    if (f1):
        sin = sin + sin1
        plot1.plot(time/1000,sin1,color='grey',linestyle='-',linewidth=1,label='100,000')
    if (f2):
        sin = sin + sin2
        plot1.plot(time/1000,sin2,color='grey',linestyle='-',linewidth=1,label='41,000')
    if (f3):
        sin = sin + sin3
        plot1.plot(time/1000,sin3,color='grey',linestyle='-',linewidth=1,label='23,000')
    if (f4):
        sin = sin + sin4
        plot1.plot(time/1000,sin4,color='grey',linestyle='-',linewidth=1,label='19,000')
    plot1.plot(time/1000,sin,color='red',linestyle='-',linewidth=2,label='Sum')
    plot1.set_xlim([tmin/1000,tmax/1000])
    plot1.set_ylim([-4,4])
    plot1.legend(loc='upper right')
    print('Calling nr_period ...')
    (px,py,pn) = nr_period(-time,sin,nmax)
    print('back')
    plot2.set_xlabel('Period [ky]')
    plot2.set_ylabel('Power spectrum')
    plot2.semilogx(1./(px*1000),py)
    plot2.set_xlim([2.,900.])
    #plot2.set_ylim([0,2])
    plot2.set_xticks([2,5,10,20,50,100,200,500])
    plt.show()
               
#plot_spectra(True,True,False,False)

# call interactive module
interactive_plot = interactive(plot_spectra,
                               f1=widgets.Checkbox(value=True,description='100,000 years',disabled=False),
                               f2=widgets.Checkbox(value=False,description='41,000 years',disabled=False),
                               f3=widgets.Checkbox(value=False,description='23,000 years',disabled=False),
                               f4=widgets.Checkbox(value=False,description='19,000 years',disabled=False))
output = interactive_plot.children[-1]
interactive_plot
```
