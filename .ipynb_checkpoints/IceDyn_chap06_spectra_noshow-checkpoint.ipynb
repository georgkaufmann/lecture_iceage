In this notebook, we read **paleo-climate** archives as time series and analyse their **spectal content**, using `python`.

We first initialize the `python` libraries.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib
import ipywidgets as widgets
```

----
## Synthetic data

Simple cosine function with arbitrary time sampling.

```python
A    = 2.
f0   = 2.  # 1/s
nin  = 150
nout = 100000

time = np.random.uniform(low=0,high=10*np.pi,size=nin)
data = A * np.cos(f0*time)
print(time.shape,time.min(),time.max())
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,4))
ax.plot(time,data,lw=0,marker='x')
ax.plot(time[np.argsort(time)],data[np.argsort(time)],lw=1)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Amplitude')
```

Calculate Lomb-Scargle periodogram 

```python
freq     = np.linspace(0.01, 10, nout)
spectrum = sp.signal.lombscargle(time, data, freq, normalize=True)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,4))
ax.plot(freq,spectrum,lw=1)
#ax.plot(time[np.argsort(time)],data[np.argsort(time)],lw=1)
ax.set_xlabel('Frequency [1/s]')
ax.set_ylabel('Spectrum')
```

## Real data

### Lomb-Scargle periodogram

```python
#-----------------------------------------------------------------------
# read data into field
#-----------------------------------------------------------------------
infile1='data/Vostok_dH.x'
#infile1='data/NGRIP_d18O.x'
time=np.loadtxt(fname=infile1,skiprows=1,usecols=0)
data=np.loadtxt(fname=infile1,skiprows=1,usecols=1)

f1 = 2
time=time/10000*3/4
data = 2*np.cos(time*f1)
print(data.mean(),data.std())
mu = data.mean()
sigma = data.std()
#data = (data - mu) / sigma
print(data.mean(),data.std(),data.min(),data.max())
```

```python
nout = 10000000
freq     = np.linspace(0.01,10, nout)
print(freq.min(),freq.max())
spectrum = sp.signal.lombscargle(time, data, freq, normalize=True)
```

```python
fig,axs = plt.subplots(2,1,figsize=(10,4),constrained_layout=True)
axs[0].plot(time,data,lw=1)
axs[0].set_xlabel('Time [a BP]')
axs[0].set_ylabel('$\delta H$ [permil]')
#axs[0].set_xlim([time.max(),time.min()])

axs[1].plot(freq,spectrum,lw=1)
axs[1].set_xlabel('Period [a]')
axs[1].set_ylabel('Norm. spectrum')
axs[1].set_yscale('log')
```

## Real data

### Spline interpolation and FFT

```python
#-----------------------------------------------------------------------
# read data into field
#-----------------------------------------------------------------------
infile1='data/Vostok_dH.x'
#infile1='data/NGRIP_d18O.x'
time=np.loadtxt(fname=infile1,skiprows=1,usecols=0)
data=np.loadtxt(fname=infile1,skiprows=1,usecols=1)

mu = data.mean()
sigma = data.std()
data = (data - mu) / sigma

spline = sp.interpolate.CubicSpline(time,data)
```

```python
tsample,dt = np.linspace(time.min(),time.max(),100001,retstep=True)

kx = np.fft.fftfreq(tsample.size, d=dt)
kx = np.fft.fftshift(kx)

F = np.fft.fft(spline(tsample))
F = np.fft.fftshift(F)

fig,axs =plt.subplots(2,1,figsize=(10,4))
axs[0].plot(tsample,spline(tsample))
axs[1].plot(1/kx,F)
#axs[1].set_xlim([0,150000])
```
