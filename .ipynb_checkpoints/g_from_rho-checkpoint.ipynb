```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile1='data/PREM.z'
rprem   = np.loadtxt(fname=infile1,skiprows=2,usecols=0)
rhoprem = np.loadtxt(fname=infile1,skiprows=2,usecols=1)
gzprem  = np.loadtxt(fname=infile1,skiprows=2,usecols=2)
```

```python
ùê∫ = 6.67408e-11
M = 5.972e24
rho0 = 13000.
rho1 = 5500.

rin  = np.arange(0,6371,1)
rout = np.arange(6371,20000,50)
gzin = 0.*rin
rho  = 0.*rin
mass = 0.
for i in range(1,len(rin)):
    mass = mass + rhoprem[i]*((rin[i]*1000)**3.-(rin[i-1]*1000)**3.)
    rho[i]  = rhoprem[i]
    gzin[i] = 4./3.*np.pi*G*mass/(rin[i]*1000)**2.
gzout = G*M/(rout*1000)**2

plt.figure(figsize=(10,6))
fig,ax1 = plt.subplots(figsize=(10,6))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,15000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color=(0.6,0.6,0.6))
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color=(0.6,0.6,0.6))
ax1.plot(rprem,gzprem,linewidth='5',color='blue',label='PREM $g$')

ax1.plot(rin,gzin,linewidth='3',color='red',label='inside $g$')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')

ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [kg/m$^3$]')
ax2.plot(rprem,rhoprem/1000,linewidth='5',linestyle='--',color='blue',label='PREM $\\rho$')
ax2.plot(rin,rho/1000,linewidth='2',color='green',label='$\\rho$')
ax2.legend(loc='lower right')
```
